```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#read data
retail_data1=pd.read_csv('Features data set.csv')
retail_data1
```

```python
#read data
retail_data2=pd.read_csv('sales data-set.csv')
retail_data2
```

```python
#read data
retail_data3=pd.read_csv('stores data-set.csv')
retail_data3
```

```python
print('feature data:' , retail_data1.shape)
print('sales data :' ,retail_data2.shape)
print('stores data:' ,retail_data3.shape)
```

```python
df=pd.merge(retail_data2,retail_data1, on=['Store','Date', 'IsHoliday'], how='left')
df=pd.merge(df,retail_data3, on=['Store'], how='left')
```

# Cleaning data

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df["MarkDown1"].fillna("0", inplace = True)
df["MarkDown2"].fillna("0", inplace = True)
df["MarkDown3"].fillna("0", inplace = True)
df["MarkDown4"].fillna("0", inplace = True)
df["MarkDown5"].fillna("0", inplace = True)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
plots = sns.lineplot(data = df, x = 'Temperature', y = 'Weekly_Sales')
```

```python
plots = sns.barplot(data = df, x = 'Type', y = 'Weekly_Sales')
```

type A in store is the most weekly sales

```python
stores_date=df.groupby(by=['Date'] , as_index=False)['Weekly_Sales'].sum()
stores_date.sort_values(by='Weekly_Sales', ascending=False)
```

most weekly sales was in date 24/12/2010	

```python
best_stores_type=df.groupby(by=['Type'] , as_index=False)['Weekly_Sales'].sum()
best_stores_type.sort_values(by='Weekly_Sales', ascending=False)
```

best weekly sales is from store of type A

```python
best_stores_sales=df.groupby(by=['Store'] , as_index=False)['Weekly_Sales'].sum()
best_stores_sales.sort_values(by='Weekly_Sales', ascending=False)
```

the best store in weekly sales was store number 20
